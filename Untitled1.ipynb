```python
# We assume campaign count greater than 32 are outliers
data3 = data2[data2['campaign']<33]
```

```python
data3.groupby(['y','campaign'],sort=True)['campaign'].count()
```

```python
# Remove ouliers in previous feature
data3.groupby(['y','previous'],sort=True)['previous'].count()
```

```python
data4 = data3[data3['previous']<31]
```

```python
data4.head()
```

```python
data4.shape
```

```python
# Handle categorical features

cat_coloumns = ['job','marital', 'education', 'contact', 'month', 'poutcome']
for col in cat_coloumns:
    data4 = pd.concat([data4.drop(col,axis=1),pd.get_dummies(data4[col],prefix=col,prefix_sep='_',
                                                             drop_first=True,dummy_na=False)],axis=1)
```

```python
#Creating a dataframe

#cat_coloumns = {'name':['job','marital', 'education', 'contact', 'month', 'poutcome']}

#data4 = pd.DataFrame(cat_coloumns, columns = ['name'])


# Creating a set of dummy variables from the gender variable

#data4 = pd.get_dummies(df['name'])

# Joining the dummy variables to the main dataframe

#data4 = pd.concat([data4], axis=1)
#print(); print(data4)
```

```python
bool_coloumns = ['housing','loan','y']
for col in bool_coloumns:
    data4[col+'_new'] = data4[col].apply(lambda x: 1 if x =='yes' else 0)
    data4.drop(col,axis=1,inplace=True)
```

```python
data4.head()
```

## Split dataset into Training set and Testing set

```python
X = data4.drop(['y_new'],axis=1)
y = data4['y_new']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
len(X_train)
```

```python
len(X_test)
```

# Model Selection

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
```

```python
from sklearn.model_selection import cross_val_score
model_score = cross_val_score(estimator=RandomForestClassifier(), X = X_train, y = y_train, cv=5)
print(model_score)
print(model_score.mean())
```

```python
from sklearn.model_selection import cross_val_score
model_score = cross_val_score(estimator=XGBClassifier(), X=X_train,y=y_train,cv=5)
print(model_score)
print(model_score.mean())
```

```python
# create parameters
model_param = {'RandomForestClassifier':{
                'model': RandomForestClassifier(),
                 'param':{
                     'n_estimators': [10, 50, 100, 150],
                     'criterion': ['gini', 'entropy'],
                     'max_depth': range(2, 4, 1),
                     'max_features': ['auto', 'log2']
                     
                 }
},
               'XGBClassifier':{
                   'model': XGBClassifier(objective='binary:logistic'),
                   'param':{
                       'learning_rate': [0.5, 0.1, 0.01, 0.001],
                       'max_depth': [3, 5, 10, 20],
                       'n_estimators': [10, 50, 100, 200]
                   }
               }
    
}
```

```python
# Gridsearch
scores = []
for model_name, mp in model_param.items():
    model_selection = GridSearchCV(estimator=mp['model'], param_grid=mp['param'],cv=5,return_train_score=True)
    model_selection.fit(X,y)
    scores.append({
        'model': model_name,
        'best_scores': model_selection.best_score_,
        'best_params': model_selection.best_params_
    })
```

```python
scores
```

```python
pd.set_option('display.max_columns', None)
best_model = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
best_model
```
